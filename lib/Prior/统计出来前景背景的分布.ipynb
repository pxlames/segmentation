```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import io
import cv2

# # 文件路径
# 读取图像和预测结果
original_image = np.array(Image.open('/home/xkw/pxlames/segmentation/img_05.png'))
prediction = np.load('/home/xkw/pxlames/segmentation/pred_05.npy')
ground_truth = np.array(Image.open('/home/xkw/pxlames/segmentation/gt_05.png'))

# 对预测结果进行阈值处理(0-1)
prediction_binary1 = (prediction > 0.1).astype(np.uint8)
prediction_binary2 = (prediction > 0.11).astype(np.uint8)
prediction_binary = np.logical_xor(prediction_binary1,prediction_binary2)
TP = np.logical_and(prediction_binary,ground_truth[:,:,0])
FP = np.logical_and(prediction_binary, np.logical_not(ground_truth[:,:,0]))

# 绘制TP和FP
plt.figure(figsize=(15, 5))

# 绘制TP
plt.subplot(121)
plt.imshow(TP, cmap='gray')
plt.title('真阳性(TP)')
plt.axis('off')

# 绘制FP 
plt.subplot(122)
plt.imshow(FP, cmap='gray')
plt.title('假阳性(FP)') 
plt.axis('off')

plt.tight_layout()
plt.show()

# 绘制原始图像、预测结果和真实标签
plt.figure(figsize=(15, 5))

# 绘制原始图像
plt.subplot(131)
plt.imshow(original_image)
plt.title('原始图像')
plt.axis('off')

# 绘制预测结果
plt.subplot(132)
plt.imshow(prediction_binary, cmap='gray')
plt.title('预测结果')
plt.axis('off')

# 绘制真实标签
plt.subplot(133)
plt.imshow(ground_truth, cmap='gray')
plt.title('真实标签')
plt.axis('off')

plt.tight_layout()
plt.show()


# 统计TP、FP处的img灰度值分布图
plt.figure(figsize=(15, 5))

# 获取TP和FP位置的灰度值
tp_values = original_image[TP]
fp_values = original_image[FP]

# 绘制TP灰度值分布直方图
plt.subplot(121)
plt.hist(tp_values.ravel(), bins=50, color='blue', alpha=0.7)
plt.title('真阳性(TP)灰度值分布')
plt.xlabel('灰度值')
plt.ylabel('频数')

# 绘制FP灰度值分布直方图
plt.subplot(122)
plt.hist(fp_values.ravel(), bins=50, color='red', alpha=0.7)
plt.title('假阳性(FP)灰度值分布')
plt.xlabel('灰度值')
plt.ylabel('频数')

plt.tight_layout()
plt.show()



# 对预测结果进行闭合操作以消除小孔洞
kernel = np.ones((8,8), np.uint8)
closed_prediction = cv2.morphologyEx(prediction_binary.astype(np.uint8), cv2.MORPH_CLOSE, kernel)

# 显示闭合操作前后的对比
plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.imshow(prediction_binary, cmap='gray')
plt.title('闭合操作前')
plt.axis('off')

plt.subplot(122)
plt.imshow(closed_prediction, cmap='gray')
plt.title('闭合操作后')
plt.axis('off')

plt.tight_layout()
plt.show()

# 更新prediction_binary为闭合后的结果
prediction_binary = closed_prediction
```
