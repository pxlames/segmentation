```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import io

# # 文件路径
# 读取图像和预测结果
original_image = np.array(Image.open('/home/xkw/pxlames/segmentation/img_05.png'))
prediction = np.load('/home/xkw/pxlames/segmentation/pred_05.npy')
ground_truth = np.array(Image.open('/home/xkw/pxlames/segmentation/gt_05.png'))

# 对预测结果进行阈值处理(0-1)
prediction_binary1 = (prediction > 0.1).astype(np.uint8)
prediction_binary2 = (prediction > 0.2).astype(np.uint8)
prediction_binary = np.logical_xor(prediction_binary1,prediction_binary2)
TP = np.logical_and(prediction_binary,ground_truth[:,:,0])
FP = np.logical_and(prediction_binary, np.logical_not(ground_truth[:,:,0]))

# 绘制TP和FP
plt.figure(figsize=(15, 5))

# 绘制TP
plt.subplot(121)
plt.imshow(TP, cmap='gray')
plt.title('真阳性(TP)')
plt.axis('off')

# 绘制FP 
plt.subplot(122)
plt.imshow(FP, cmap='gray')
plt.title('假阳性(FP)') 
plt.axis('off')

plt.tight_layout()
plt.show()

# 绘制原始图像、预测结果和真实标签
plt.figure(figsize=(15, 5))

# 绘制原始图像
plt.subplot(131)
plt.imshow(original_image)
plt.title('原始图像')
plt.axis('off')

# 绘制预测结果
plt.subplot(132)
plt.imshow(prediction_binary, cmap='gray')
plt.title('预测结果')
plt.axis('off')

# 绘制真实标签
plt.subplot(133)
plt.imshow(ground_truth, cmap='gray')
plt.title('真实标签')
plt.axis('off')

plt.tight_layout()
plt.show()
```
